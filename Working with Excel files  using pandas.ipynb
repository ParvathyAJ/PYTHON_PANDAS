# EXPORT EXCEL IN PANDAS

```python
# Start with importing the baby_names csv

url = "https://data.cityofnewyork.us/resource/25th-nujf.csv"
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
# Split them into girls and boys and write them in the excel file using pandas
```

```python
girls = baby_names[baby_names["gndr"] == "FEMALE"]
boys = baby_names[baby_names["gndr"] == "MALE"]
```

```python
boys
```

```python
excel_file = pd.ExcelWriter("Baby_Names.xlsx")
```

```python
girls.to_excel(excel_file, sheet_name = "Girls", index = False, encoding="utf-8")
```

```python
boys.to_excel(excel_file, sheet_name = "Boys", index = False, columns = ["brth_yr", "gndr", "ethcty"],encoding="utf-8")
```

```python
excel_file.save()
```
