```python
bb = data.DataReader(name = "BB", data_source="stooq", start="2007-1-1", end="2020-12-31")
bb.head(3)
```

```python
bb.plot()
bb.plot(y ="Volume")
bb.plot(y ="High")
```

```python
def rank_perf(stock):
    if stock <= 10:
        return "Poor"
    elif stock <=50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
bb["Close"].apply(rank_perf)
```

```python
bb["Close"].apply(rank_perf).value_counts()
```

```python
plt.style.use("classic")
bb["Close"].apply(rank_perf).value_counts().plot()
```

```python
bb["Close"].apply(rank_perf).value_counts().plot(kind = "barh")
```

```python
bb["Open"].mean()
```

```python
def rek(s):
    if s > 29.50214320567376:
        return "Above Average"
    else:
        return "Below Average"
```

```python
bb["Open"].apply(rek).value_counts()
```

```python
plt.style.use("dark_background")
```

```python
bb["Open"].apply(rek).value_counts().plot(kind = "pie", legend = True)
```
