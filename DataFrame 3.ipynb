```python
# Apply method on columns


bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_million(number):
    return str(number) + "MILLION"

bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_add_million)
```

```python
bond
```

```python
columns =['Budget', 'Box Office', 'Bond Actor Salary']

for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_million)
    
```

```python
bond
```

```python
#  Apply method on rows

bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(rows):
    actor = rows[1]
    budget = rows[4]
    
    if actor == "Sean Connery":
        return "The Best!"
    elif actor =="Roger Moore" and budget >80:
        return "Enjoyable!"
    else:
        return "No Clue!"
    
bond.apply(good_movie, axis = "columns")    
```
