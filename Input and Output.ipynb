# Export CSV



```python
url = "https://data.cityofnewyork.us/resource/25th-nujf.csv"
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
baby_names.to_csv("NYC Baby Names.csv", index = False, columns =[ "nm","cnt"], encoding ="utf-8")
```
