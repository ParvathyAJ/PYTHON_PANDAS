# OUTER JOIN

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
food = pd.read_csv("Restaurant - Foods.csv")
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes=[" - Week1", " - Week2"], indicator= True)
```

# left_on and right_on Parameters


```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
food = pd.read_csv("Restaurant - Foods.csv")
customer = pd.read_csv("Restaurant - Customers.csv")
```

```python
week2.head(2)
```

```python
customer.head(2)
```

```python
#  Column names customer id and id are same but the labels are different, we can use the on paramater 
#  only if they match in their labels
```

```python
week2.merge(customer, how ="left", left_on =  "Customer ID", right_on = "ID")
```

```python
week2.merge(customer, how ="left", left_on =  "Customer ID", right_on = "ID", sort = True).drop("ID", axis=1)
```

# Merging with left_index and right_index Parameters

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
food = pd.read_csv("Restaurant - Foods.csv", index_col="Food ID")
customer = pd.read_csv("Restaurant - Customers.csv", index_col="ID")
```

```python
food.head(3)
```

```python
customer.head(3)
```

```python
week1.head(3)
```

```python
sales = week1.merge(customer, how ="left", left_on = "Customer ID", right_index = True)
```

```python
sales.head(2)
```

```python
sales.merge(food, how = "left", left_on = "Food ID", right_index = True)
sales.head(5)
```
